```python
%sql sqlite:///jupyter_sql_tutorial.db
```

```python
%%sql
SELECT *
FROM sale

```

```python
%%sql
SELECT count(*)
FROM sale

```

```python
%%sql
SELECT 
    
    SUM(Total_Amount) AS monthly_revenue
FROM 
    sale

```

```python
%%sql

SELECT 
    Product_Category,
    SUM(Quantity) AS total_quantity_sold,
    SUM(Total_Amount) AS total_revenue
FROM 
    sale
GROUP BY 
    Product_Category
ORDER BY 
    total_quantity_sold DESC;

```

```python
%%sql
SELECT 
    Gender,
    SUM(Total_Amount) AS total_spent,
    COUNT(*) AS total_purchases
FROM 
    sale
GROUP BY 
    Gender
ORDER BY 
    total_spent DESC;
```

```python
%%sql

SELECT DISTINCT Gender, Age FROM sale;
```
