```python
import pandas as pd
import numpy as np
df=pd.read_csv("Titanic.csv")
df
```

```python
df = df.drop('Cabin', axis=1)
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Age', 'Embarked'])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Bar plot for passenger demographics (Sex)
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=df)
plt.title('Number of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


```

```python
# Scatter plot for Fare vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Fare vs. Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

```

```python
# Bar plot for survival rates by class
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

# Descriptive Captions
Number of Passengers by Sex: This bar plot shows the distribution of passengers by sex. It helps us understand the gender demographics of the passengers.
 
Age Distribution of Passengers: This histogram displays the age distribution of passengers, providing insights into the age range and frequency.

Fare vs. Age: This scatter plot illustrates the relationship between passenger age and ticket fare, highlighting any potential patterns or outliers.

Survival Rates by Passenger Class: This bar plot compares the survival rates across different passenger classes, showing how class affected survival chances.

```python
data=pd.read_csv("covid19_us_county.csv")
data
```

```python
data.count()
```

```python

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter data for specific countries
countries = ['USA', 'India', 'Brazil', 'UK']
filtered_data = data[data['country'].isin(countries)]

# Plot daily new cases
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = filtered_data[filtered_data['country'] == country]
    plt.plot(country_data['date'], country_data['new_day_cases'], label=country)
plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.show()

# Plot daily new deaths
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = filtered_data[filtered_data['country'] == country]
    plt.plot(country_data['date'], country_data['new_day_deaths '], label=country)
plt.title('Daily New COVID-19 Deaths')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.legend()
plt.show()


```

```python

# Plot daily new cases
plt.figure(figsize=(10, 5))
plt.plot(data.index, data['new_day_cases'], label='Daily New Cases', color='blue')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Daily New COVID-19 Cases')
plt.legend()
plt.show()

# Plot daily deaths
plt.figure(figsize=(10, 5))
plt.plot(data.index, data['new_day_deaths'], label='Daily Deaths', color='red')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.title('Daily COVID-19 Deaths')
plt.legend()
plt.show()


```
